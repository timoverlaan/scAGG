```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("out/results/gsea/gsea_all_cts.csv")
df.drop(columns=["Unnamed: 0", "Name"], inplace=True)

col_db = []
col_term = []
for term in df["Term"].values:
    db, term = term.split("__")
    col_db.append(db.replace("_", " "))
    col_term.append(term.replace("_", " "))
df["Database"] = col_db
df["Term"] = col_term

df
```

```python
df["color_nes"]
```

```python
COLOR_COL = "ES"
COLOR_MAP = plt.cm.coolwarm

# Scale the color values
color_col = df[COLOR_COL]
color_col_norm = (color_col - color_col.min()) / (color_col.max() - color_col.min())
def rgba_to_hex(rgba_color):
    return '#%02x%02x%02x' % tuple(int(x * 255) for x in rgba_color[:3])
df["color"] = [rgba_to_hex(c) for c in COLOR_MAP(color_col_norm)]

fig, ax = plt.subplots(figsize=(3, 6))

df.sort_values(by="Database", ascending=False, inplace=True)

CELLTYPES = [ct for ct in sorted(df["Celltype"].unique())]
TERMS = df["Term"].unique()

for i, celltype in enumerate(CELLTYPES):
    for j, term in enumerate(TERMS):
        row = df[(df["Celltype"] == celltype) & (df["Term"] == term)]
        if len(row["NES"].values) == 0:
            continue

        # coding:
        # - marker: database ("s" for KEGG, "o" for GO)
        # - color: NES value
        # - size: p-value

        db = row["Database"].values[0]
        color = row["color"].values[0]
        pval = row["FDR q-val"].values[0]

        if "KEGG" in db:
            marker = "s"
        elif "GO" in db:
            marker = "o"

        # Make marker size proportional to -log(pval)
        # Scale to [5, 25]
        markersize = - np.log(pval + 0.1)
        all_markersizes = - np.log(df["FDR q-val"].values + 0.1)
        markersize = (markersize - all_markersizes.min()) / (all_markersizes.max() - all_markersizes.min())
        markersize = 5 + 20 * markersize
        print(markersize)

        ax.plot(i, j, marker=marker, color=color, markersize=markersize)

ax.set_xlim(-0.5, len(CELLTYPES) - 0.5)

CELLTYPES = [ct.replace("_", " ") for ct in CELLTYPES]
TERMS = [term.split(" (GO")[0] for term in TERMS]
ax.set_xticks(range(len(CELLTYPES)), CELLTYPES, rotation=45, ha='right')
ax.set_yticks(range(len(TERMS)), TERMS)

# Legends
marker_legend_elements = [
    plt.plot([], [], marker='o', color='gray', linestyle='none', 
             label='GO', markersize=10)[0],
    plt.plot([], [], marker='s', color='gray', linestyle='none', 
             label='KEGG', markersize=10)[0]
]
pvals_legend = [0.05, 0.04, 0.03, 0.02, 0.01]
marker_size_legend_elements = [
    # plt.plot([], [], marker='o', color='gray', linestyle='none', 
    #          label='FDR q-val = 0.05', markersize=5)[0],
    # plt.plot([], [], marker='o', color='gray', linestyle='none', 
    #          label='FDR q-val = 0.01', markersize=10)[0],
    # plt.plot([], [], marker='o', color='gray', linestyle='none', 
    #          label='FDR q-val = 0.001', markersize=20)[0]
]
for pval in pvals_legend:
    markersize = - np.log(pval + 0.1)
    all_markersizes = - np.log(df["FDR q-val"].values + 0.1)
    markersize = (markersize - all_markersizes.min()) / (all_markersizes.max() - all_markersizes.min())
    markersize = 5 + 20 * markersize
    marker_size_legend_elements.append(
        plt.plot([], [], marker='o', color='gray', linestyle='none', 
                 label=f'p = {pval}', markersize=markersize)[0]
    )

leg = ax.legend(handles=marker_legend_elements, loc='upper right', title='Database', bbox_to_anchor=(2.1, 0.7))
plt.gca().add_artist(leg)
ax.legend(handles=marker_size_legend_elements, loc='upper right', title='p-val', bbox_to_anchor=(2.1, 1))

# Colorbar
sm = plt.cm.ScalarMappable(cmap=COLOR_MAP, norm=plt.Normalize(vmin=df[COLOR_COL].min(), vmax=df[COLOR_COL].max()))
cbar = fig.colorbar(sm, ax=ax, fraction=0.05, pad=0.1, aspect=25)
cbar.set_label("Enrichment Score", rotation=90, labelpad=5)

ax.grid(False)
plt.show()
```
